```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
from matplotlib.patches import Rectangle
from matplotlib.collections import PatchCollection
import numpy as np
import pandas as pd
import watermark
import seaborn as sns
import scipy as sp

%load_ext watermark
%matplotlib inline
```

```python
data = pd.read_csv('./movies.csv')
data
```

```python
data2 = pd.DataFrame(data)
data2
```

```python
data10 = data2[(data2.Distributor=="MGM")|
               (data2.Distributor=="Sony Pictures")|
              (data2.Distributor=="Universal")|
              (data2.Distributor=="Walt Disney Pictures")|
              (data2.Distributor=="Warner Bros")|
              (data2.Distributor=="20th Century Fox")|
              (data2.Distributor=="Paramount Pictures")|
              (data2.Distributor=="Miramax")
              ]
data10
```

```python
data31 = pd.concat([data10.iloc[:,4], data10.iloc[:,8]], axis=1)
data31
```

```python
data32 = data31.replace('Unknown',np.NaN)
data33 = data32.replace('0',np.NaN)
```

```python
data10 = data33.dropna(axis=0, how='any')
data10
```

```python
sns.swarmplot(x='Distributor',y='Production Budget', data=data10)
plt.xticks(rotation=90) 
```

```python
ax = sns.violinplot(x="Distributor", y="Production Budget", data=data10, inner=None)
ax = sns.swarmplot(x="Distributor", y="Production Budget", data=data10,
                   color="white", edgecolor="gray")
plt.xticks(rotation=90)
```

```python
data42 = data2[(data2.Distributor=="MGM")]
data42
```

```python
data43 = pd.concat([data42.iloc[:,2], data42.iloc[:,13],data42.iloc[:,14]], axis=1)
data43
```

```python
data44 = data43.replace('Unknown',np.NaN)
data45 = data44.replace('0',np.NaN)
data46 = data45.dropna(axis=0, how='any')
data46
```

```python
sns.pairplot(data46)
```

```python

ax = sns.kdeplot(data46['Worldwide Gross'], data46['IMDB Rating'],
                  cmap="Reds", shade=True, shade_lowest=False)
```

```python

ax = sns.kdeplot(data46['Worldwide Gross'], data46['Rotten Tomatoes Rating'],
                  cmap="Reds", shade=True, shade_lowest=False)
```

```python
cols = ['Worldwide Gross', 'Rotten Tomatoes Rating', 'IMDB Rating']
pp = sns.pairplot(data46[cols], height=1.8, aspect=1.8, 
                  
                  plot_kws=dict(edgecolor="black", linewidth=0.5))
fig = pp.fig 
fig.subplots_adjust(top=0.93, wspace=0.3)
```

```python
fig = plt.figure(figsize=(8, 6))
ax = fig.add_subplot(111, projection='3d')

xs = data46['Worldwide Gross']
ys = data46['Rotten Tomatoes Rating']
zs = data46['IMDB Rating']
ax.scatter(xs, ys, zs, s=50, alpha=0.6, edgecolors='w')

ax.set_xlabel('Worldwide Gross')
ax.set_ylabel('Rotten Tomatoes Rating')
ax.set_zlabel('IMDB Rating')
```
